```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
image=cv2.imread('sheets.jpg',0)
    
```

```python
def display(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
```

```python

display(image)
```

```python
#median_blurring


median_img=cv2.medianBlur(image,7)
```

```python
display(median_img)
```

```python
#gaussian_blurring
blurred=cv2.GaussianBlur(image,(5,5),10)
```

```python
display(blurred)
```

```python
#calculating x-gradiant for initial image
sobelx=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)
```

```python
display(sobelx)
```

```python
#calculating x-gradiant for median-blurred image
sobelx1=cv2.Sobel(median_img,cv2.CV_64F,1,0,ksize=5)
```

```python
display(sobelx1)
```

```python
#calculating x-gradient for gaussian blurred image
sobelx2=cv2.Sobel(blurred,cv2.CV_64F,1,0,ksize=5)
```

```python
display(sobelx2)
```

```python
sobelx.max()
```

```python
sobelx1.max()
```

```python
sobelx2.max()
```

```python
# now we calculate threshold 
#thresold for median_blurred image

value,thres=cv2.threshold(sobelx1,300,5769,cv2.THRESH_BINARY_INV)
```

```python
display(thres)
```

```python
#threshold for gaussian_blurred image
value,thres1=cv2.threshold(sobelx2,600,4209,cv2.THRESH_BINARY_INV)
```

```python
display(thres1)
```

```python
#theshold for initial image

value,thres=cv2.threshold(sobelx,800,8724,cv2.THRESH_BINARY_INV)
```

```python
display(thres)
```
